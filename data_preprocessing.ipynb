```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import joblib
import torch
df = pd.read_csv('211210-DRM-total.csv')
```

```python
mt_cols = df.iloc[:,2:25].columns
sp_cols = df.iloc[:,25:70].columns
pt_cols = df.iloc[:,70:76].columns
op_cols = df.iloc[:,76:].columns
```

### sklearn을 활용해 train,val,test로 나눠서 구한다.
#### 0.6,0.2,0.2로 나눠서 구한다.

```python
from sklearn.model_selection import train_test_split
train_df,test_df = train_test_split(df,test_size = 0.4,random_state = 21)
val_df,test_df = train_test_split(test_df,test_size = 0.5, random_state=21)
```

```python
scaler_op = MinMaxScaler()
scaler_mt = MinMaxScaler()
scaler_sp = MinMaxScaler()
scaler_pt = MinMaxScaler()
```

### train 학습하는 방법

```python
train_df[op_cols] = scaler_op.fit_transform(train_df[op_cols])
train_df[mt_cols] = scaler_mt.fit_transform(train_df[mt_cols])
train_df[sp_cols] = scaler_sp.fit_transform(train_df[sp_cols])
train_df[pt_cols] = scaler_pt.fit_transform(train_df[pt_cols])
```

### validation 학습하는 방법

```python
val_df[op_cols] = scaler_op.transform(val_df[op_cols])
val_df[mt_cols] = scaler_mt.transform(val_df[mt_cols])
val_df[sp_cols] = scaler_sp.transform(val_df[sp_cols])
val_df[pt_cols] = scaler_pt.transform(val_df[pt_cols])
```

### test 학습하는 방법

```python
test_df[op_cols] = scaler_op.transform(test_df[op_cols])
test_df[mt_cols] = scaler_mt.transform(test_df[mt_cols])
test_df[sp_cols] = scaler_sp.transform(test_df[sp_cols])
test_df[pt_cols] = scaler_pt.transform(test_df[pt_cols])
```

### 따로 scaler 저장

```python
joblib.dump(scaler_mt, 'scaler/scaler_mt.pkl')
joblib.dump(scaler_sp,'scaler/scaler_sp.pkl')
joblib.dump(scaler_pt,'scaler/scaler_pt.pkl')
joblib.dump(scaler_op,'scaler/scaler_op.pkl')
```
